```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk import pos_tag
```

```python
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
text = """
Natural Language Processing is an important field of AI.
We are learning different NLP techniques in universities.
Researchers develop intelligent systems to understand human language.
"""

print("Original Text:\n")
print(text)
```

```python
sentences = sent_tokenize(text)

print("Sentence Tokenization:\n")

for i, sentence in enumerate(sentences):
    print("Sentence", i+1, ":", sentence)
```

```python
words = word_tokenize(text)

print("Word Tokens:\n")
print(words)
```

```python
pos_tags = pos_tag(words)

print("Part of Speech Tags:\n")
print(pos_tags)
```

```python
print("Word  ->  POS Tag\n")

for word, tag in pos_tags:
    print(word, "->", tag)
```

```python
print("POS Tagging for Each Sentence:\n")

for sentence in sentences:
    words = word_tokenize(sentence)
    tagged_words = pos_tag(words)
    print(tagged_words)
    print()
```

```python
print("Common POS Tags Meaning:\n")

print("NN  -> Noun")
print("NNS -> Plural Noun")
print("NNP -> Proper Noun")
print("VB  -> Verb Base Form")
print("VBD -> Verb Past Tense")
print("VBG -> Verb Present Participle")
print("JJ  -> Adjective")
print("RB  -> Adverb")
print("DT  -> Determiner")
```

```python
# Dictionary to convert POS tags into readable form

pos_dictionary = {
    "NN": "Noun",
    "NNS": "Plural Noun",
    "NNP": "Proper Noun",
    "NNPS": "Proper Noun (Plural)",
    "VB": "Verb",
    "VBD": "Verb (Past Tense)",
    "VBG": "Verb (Gerund)",
    "VBN": "Verb (Past Participle)",
    "VBP": "Verb (Present)",
    "VBZ": "Verb (Present Singular)",
    "JJ": "Adjective",
    "JJR": "Adjective (Comparative)",
    "JJS": "Adjective (Superlative)",
    "RB": "Adverb",
    "RBR": "Adverb (Comparative)",
    "RBS": "Adverb (Superlative)",
    "DT": "Determiner",
    "IN": "Preposition",
    "PRP": "Pronoun",
    "CC": "Conjunction"
}

print("Word = Part of Speech\n")

for word, tag in pos_tags:
    meaning = pos_dictionary.get(tag, "Other")
    print(word, "=", meaning)
```
