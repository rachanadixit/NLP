```python
import kagglehub
import os
from nltk.util import ngrams
from collections import Counter, defaultdict
```

```python
path = kagglehub.dataset_download("gupta24789/text8-word-embedding")

file_path = os.path.join(path, "text8")
```

```python
import re

with open(file_path, 'r', encoding='utf-8') as f:
    text = f.read()[:300000]   # limit size for speed

text = text.lower()

text = re.sub(r'[^a-z\s]', '', text)

words = text.split()

print("Sample words:", words[:20])
```

```python
N = 2
```

```python
ngram_counts = defaultdict(Counter)

for i in range(len(words) - N + 1):
    context = tuple(words[i:i + N - 1])   # (N-1 words)
    next_word = words[i + N - 1]

    ngram_counts[context][next_word] += 1
```

```python
ngram_prob = {}

for context, counter in ngram_counts.items():
    total_count = sum(counter.values())

    ngram_prob[context] = {
        word: count / total_count
        for word, count in counter.items()
    }
```

```python
def predict_next_word(input_sequence):
    input_tokens = input_sequence.lower().split()


    context = tuple(input_tokens[-(N - 1):])

    if context not in ngram_prob:
        print("\nNo prediction available for:", context)
        return

    next_words = ngram_prob[context]

    predicted_word = max(next_words, key=next_words.get)

    print("\nInput:", input_sequence)
    print("Context:", context)
    print("Predicted Next Word:", predicted_word)
    print("Probability:", next_words[predicted_word])
```

```python
predict_next_word("king")
predict_next_word("natural")
predict_next_word("language")
predict_next_word("science")
```
