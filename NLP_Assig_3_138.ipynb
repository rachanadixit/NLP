N-gram Language Model for Next Word Prediction


```python
import nltk
from nltk.util import ngrams
from nltk.tokenize import word_tokenize
from collections import Counter, defaultdict
```

```python
nltk.download('punkt')
```

```python
text = """
Natural language processing is an important field of artificial intelligence.
Natural language processing helps computers understand human language.
We are learning natural language processing techniques.
"""

print("Original Text:\n")
print(text)
```

```python
nltk.download('punkt_tab')
tokens = word_tokenize(text.lower())

print("Tokenized Words:\n")
print(tokens)
```

```python
bigrams = list(ngrams(tokens, 2))

print("Generated Bigrams:\n")
print(bigrams)
```

```python
bigram_freq = Counter(bigrams)

print("Bigram Frequencies:\n")

for bigram, freq in bigram_freq.items():
    print(bigram, ":", freq)
```

```python
model = defaultdict(list)

for w1, w2 in bigram_freq:
    model[w1].append((w2, bigram_freq[(w1, w2)]))

print("Language Model:\n")
print(dict(model))
```

```python
def predict_next_word(word):

    if word not in model:
        print("Word not found in corpus")
        return

    next_words = model[word]

    # select the most frequent next word
    predicted = max(next_words, key=lambda x: x[1])

    print("Input Word:", word)
    print("Predicted Next Word:", predicted[0])
```

```python
model = defaultdict(list)

for (w1, w2), count in bigram_freq.items():
    model[w1].append((w2, count))

# Convert to probabilities
prob_model = {}

for w1, next_words in model.items():
    total = sum(count for _, count in next_words)
    prob_model[w1] = [(w2, count / total) for w2, count in next_words]
```

```python
def predict_next_word(word):
    if word not in prob_model:
        print("Word not found in corpus")
        return

    next_words = prob_model[word]

    predicted = max(next_words, key=lambda x: x[1])

    print("Input Word:", word)
    print("Predicted Next Word:", predicted[0])
    print("Probability:", predicted[1])
```

```python
predict_next_word("natural")
predict_next_word("language")
predict_next_word("processing")
```
