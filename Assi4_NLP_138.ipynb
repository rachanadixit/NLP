```python
import kagglehub

# Download dataset again
path = kagglehub.dataset_download("gupta24789/text8-word-embedding")

print("Dataset path:", path)
```

```python
import os

file_path = os.path.join(path, "text8")

with open(file_path, 'r', encoding='utf-8') as f:
    text = f.read()

print("Loaded successfully!")
print(text[:200])
```

```python
words = text.split()

sentence_length = 50
sentences = [
    words[i:i + sentence_length]
    for i in range(0, len(words), sentence_length)
]

print("Sample sentence:", sentences[0])
```

```python
from gensim.models import Word2Vec

sentences = sentences[:30000]

model = Word2Vec(
    sentences=sentences,
    vector_size=100,
    window=5,
    min_count=5,
    workers=4,
    epochs=5
)
```

```python

print("\nWords similar to 'king':")
print(model.wv.most_similar("king"))

print("\nSimilarity between 'king' and 'queen':")
print(model.wv.similarity("king", "queen"))

print("\nAnalogy: king - man + woman = ?")
print(model.wv.most_similar(positive=['king', 'woman'], negative=['man']))
```

```python
model.save("word2vec_text8.model")

print("\nModel saved successfully!")
```
