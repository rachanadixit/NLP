Implement Tokenization, Stemming, and Lemmatization on a Small Corpus


```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
corpus = """
Natural Language Processing is an important field of AI.
It helps computers understand and interpret human language.
Many researchers are working on improving NLP systems.
we are learning different NLP techniques in universities.
"""
print(corpus)
```

```python
nltk.download('punkt_tab')
sentences = sent_tokenize(corpus)

print("Sentence Tokenization:\n")

for i, sentence in enumerate(sentences):
    print("Sentence", i+1, ":", sentence)
```

```python
words = word_tokenize(corpus)

print("Word Tokenization:\n")
print(words)
```

```python
stemmer = PorterStemmer()

stemmed_words = []

for word in words:
    stem_word = stemmer.stem(word)
    stemmed_words.append(stem_word)

print("Stemmed Words:\n")
print(stemmed_words)
```

```python
print("Original Word -> Stemmed Word\n")

for word in words:
    print(word, "->", stemmer.stem(word))
```

```python
lemmatizer = WordNetLemmatizer()

lemmatized_words = []

for word in words:
    lemma_word = lemmatizer.lemmatize(word)
    lemmatized_words.append(lemma_word)

print("Lemmatized Words:\n")
print(lemmatized_words)
```

```python
print("Original Word -> Lemmatized Word\n")

for word in words:
    print(word, "->", lemmatizer.lemmatize(word))
```

```python
print("Word | Stem | Lemma")
print("-----------------------")

for word in words:
    stem = stemmer.stem(word)
    lemma = lemmatizer.lemmatize(word)
    print(word, "|", stem, "|", lemma)
```
